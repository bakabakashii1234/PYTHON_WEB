```python
import turtle
#tạo con trỏ
t = turtle.Turtle()
#define pen size
t.pensize(2)
#define pen color
t.pencolor("black") 
#Vẽ mặt trời
sun_size = 50
t.fillcolor ("yellow")
t.penup()
t.goto(100, 100)
t.pendown()
t.begin_fill()
t.circle(sun_size)
t.end_fill()
#vẽ tia của mặt trời
#tia dai 1
t.right(90)
t.forward(40)
#tia dai 2
t.penup()
t.goto(50, 150)
t.right(90)
t.pendown()
t.forward(40)
#tia dai 3
t.penup()
t.goto(100, 200)
t.right(90)
t.pendown()
t.forward(40)
#tia dai 4
t.penup()
t.goto(150, 150)
t.right(90)
t.pendown()
t.forward(40)
#tia ngan 1
t.penup()
t.goto(135, 115)
t.right(45)
t.pendown()
t.forward(30)
#tia ngan 2
t.penup()
t.goto(65, 115)
t.right(90)
t.pendown()
t.forward(30)
#tia ngan 3
t.penup()
t.goto(65, 185)
t.right(90)
t.pendown()
t.forward(30)
#tia ngan 4
t.penup()
t.goto(135, 185)
t.right(90)
t.pendown()
t.forward(30)
#vẽ tam giac

turtle.mainloop()
```
